```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()

base_url = str("https://fminside.net/clubs")
driver.get(base_url)

# fm21 datası çekmek için kod #
#Select(driver.find_element_by_xpath("//*[@id='sidebar']/div[2]/form/div[1]/select")).select_by_index(1)

dict_name_link = {"CName": [], "CLink": []}

# "CRating": [], "CLeague": [], "CBalance": [],"CBudget": [], "CTraining_Facilities": [], "CYouth_Facilities": []

for i in range(99):
    loadmore = driver.find_element_by_xpath("//*[@id='club_table']/div/div[2]/a") 
    driver.execute_script("arguments[0].click();", loadmore)
    time.sleep(2)
    
    
    
cnames_links = driver.find_elements_by_xpath("//span[@class='name']/b/a")

for cname_link in cnames_links:
    dict_name_link["CName"].append(cname_link.text)
    dict_name_link["CLink"].append(cname_link.get_attribute('href'))

```

```python
df_temp = pd.DataFrame(dict_name_link)
df_temp
```

```python
df_temp.to_csv("fm21_clubs_links.csv", index=False)
```
