```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np
import time
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

base_url = str("https://fminside.net/players")
driver.get(base_url)

dict_name_link = {"Name": [], "Link": [], "Position": []}


for i in range(19):
    loadmore = driver.find_element_by_xpath("//*[@id='player_table']/div/div[3]/a")
    driver.execute_script("arguments[0].click();", loadmore)
    time.sleep(1)
    
    
names_links = driver.find_elements_by_xpath("//span[@class='name']/b/a")
positions = driver.find_elements_by_xpath("//span[@class='desktop_positions']")
for name_link in names_links:
    dict_name_link["Name"].append(name_link.text)
    dict_name_link["Link"].append(name_link.get_attribute('href'))
    
for position in positions:
    dict_name_link["Position"].append(position.text)
```

```python
df_temp = pd.DataFrame(dict_name_link)
df_temp
```

```python
df_temp = df_temp[df_temp["Position"] != "GK"]
df_temp
```

```python
df_temp.to_csv("fm22_clubs_links.csv", index=False)
```
